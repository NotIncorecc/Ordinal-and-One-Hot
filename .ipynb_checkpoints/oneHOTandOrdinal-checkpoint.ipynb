```python
def ordinal(data,feature):
    ordinals = np.array(data[feature])
    indexMap = np.unique(ordinals)
    ordinalMap = pd.Series(range(1,len(indexMap)+1), index=indexMap)
    encoded = ordinalMap.loc[ordinals]
    data[feature+'_encoded'] = np.array(encoded)
```

```python
ordinal(student_data, 'Section')
```

```python
def oneHOT(data,feature):
    values = np.array(data[feature])
    indexMap = np.unique(values)
    listOfoneHOTextentions = [np.array([1 if x==val else 0 for x in data[feature]]) for val in indexMap]
    #print(listOfoneHOTextentions)
    for i,encoded_HOT in zip(indexMap,listOfoneHOTextentions):
        data[i+'_HOT'] =  encoded_HOT
```

```python
oneHOT(student_data, 'House Color')
```
